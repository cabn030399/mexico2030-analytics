```python
mexico_matches = df[(df["home_team"] == "Mexico") |
                    (df["away_team"] == "Mexico")]
mexico_matches.head()
```

```python
mexico_matches.shape
```

primer y ultimo partido de mexico en el dataset y lapzo de tiempo entre primer y ultimo registro disponible

```python
mexico_matches.sort_values("date").head(1)
mexico_matches.sort_values("date").tail(1)
print("Primer partido:", mexico_matches["date"].min())
print("Último partido:", mexico_matches["date"].max())
first_match = mexico_matches.sort_values("date").iloc[0]

print("Fecha:", first_match["date"])
print("Local:", first_match["home_team"])
print("Visitante:", first_match["away_team"])
print("Marcador:",
      f'{int(first_match["home_score"])} - {int(first_match["away_score"])}')
print("Torneo:", first_match["tournament"])
print("Ciudad:", first_match["city"])
print("País:", first_match["country"])
last_match = mexico_matches.sort_values("date").iloc[-1]

print("Fecha:", last_match["date"])
print("Local:", last_match["home_team"])
print("Visitante:", last_match["away_team"])
print("Marcador:",
      f'{int(last_match["home_score"])} - {int(last_match["away_score"])}')
print("Torneo:", last_match["tournament"])
print("Ciudad:", last_match["city"])
print("País:", last_match["country"])
mexico_matches["date"] = pd.to_datetime(mexico_matches["date"])
first_date = mexico_matches["date"].min()
last_date = mexico_matches["date"].max()

difference = last_date - first_date

print("Primer partido:", first_date.date())
print("Último partido:", last_date.date())
print("Días transcurridos:", difference.days)
print("Años aproximados:", round(difference.days / 365.25, 2))
```

```python
#Resumen ejecutivo 
first_match = mexico_matches.sort_values("date").iloc[0]
last_match = mexico_matches.sort_values("date").iloc[-1]

years = round(
    (pd.to_datetime(last_match["date"]) -
     pd.to_datetime(first_match["date"])).days / 365.25,
    2
)

print("========== RESUMEN HISTÓRICO MÉXICO ==========")
print(f'Primer partido registrado: {first_match["date"]}')
print(f'{first_match["home_team"]} {int(first_match["home_score"])} - {int(first_match["away_score"])} {first_match["away_team"]}')
print()
print(f'Último partido registrado: {last_match["date"]}')
print(f'{last_match["home_team"]} {int(last_match["home_score"])} - {int(last_match["away_score"])} {last_match["away_team"]}')
print()
print(f'Años cubiertos por el dataset: {years}')
```
